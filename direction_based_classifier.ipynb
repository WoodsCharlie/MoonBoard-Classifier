```python
import numpy as np
import pandas as pd
DIMS = (18,11)
```

```python
cols = []
for i in range(DIMS[0]):
    for j in range(DIMS[1]):
        cols.append((i,j))
df = pd.read_csv('datasets/flattenedClimbs.csv', header=None).astype(bool)
df.columns = cols
df
```

```python
df['holds'] = df.apply(lambda row: list(row[row].index), axis=1)
df
```

```python
grades = pd.read_csv('datasets/grades.txt', names=['grade'])
grades
df = pd.concat((df, grades), axis=1, sort=False)
df
```

```python
def find_orange(hld, hlds):
    i = hld[0]
    hold = hld[1:]
    below = hlds[i+1:,1:]
    j = i-1
    while j >= 0 and hlds[j][1] == hold[0]:
        below = np.concatenate(([hlds[j,1:]], below))
        j -= 1
    if below.shape[0] == 0:
        return np.array([np.nan, np.nan])
    dists = np.sqrt((hold[0]-below[:,0])**2+(hold[1]-below[:,1])**2)
    weights = 1/dists
    orange = np.dot(weights, below)/(weights.sum())
    return orange
```

```python
def find_quad(row):
    slope = row[4]
    if slope == 0:
        if row[3] > row[1]:
            return 3
        else:
            return 0
    if slope == np.inf or slope == -np.inf:
        return 1
    if slope > 0:
        if slope <= 1:
            return 0
        else:
            return 1
    else:
        if slope <= -1:
            return 2
        else:
            return 3
```

```python
def for_each_climb(row):
    orig = np.array(row.holds)
    grade = row.grade
    idx = np.arange(orig.shape[0]).reshape(orig.shape[0],1)
    holds = np.concatenate((idx, orig), axis=1)
    orange = np.apply_along_axis(find_orange, 1, holds, holds)
    hold_orange = np.concatenate((orig, orange), axis=1)
    slopes = -(hold_orange[:,2]-hold_orange[:,0])/(hold_orange[:,3]-hold_orange[:,1])
    hold_slope = np.concatenate((hold_orange, slopes.reshape(slopes.shape[0], 1)), axis=1)
    quads = np.apply_along_axis(find_quad, 1, hold_slope)
    hold_quad = np.concatenate((orig, quads.reshape(slopes.shape[0], 1)), axis=1)
    for row in hold_quad:
        master[row[0],row[1],row[2],0] += grade
        master[row[0],row[1],row[2],1] += 1
```

```python
master = np.zeros((18,11,4,2))
df.apply(for_each_climb, axis=1)
```

```python
classifier = master[:,:,:,0]/master[:,:,:,1]
classifier.shape
```

```python
classifier
```

```python
overall_mean = np.nanmean(classifier)
overall_mean
```

```python
def classify_climb(row, fn):
    orig = np.array(row.holds)
    idx = np.arange(orig.shape[0]).reshape(orig.shape[0],1)
    holds = np.concatenate((idx, orig), axis=1)
    orange = np.apply_along_axis(find_orange, 1, holds, holds)
    hold_orange = np.concatenate((orig, orange), axis=1)
    slopes = -(hold_orange[:,2]-hold_orange[:,0])/(hold_orange[:,3]-hold_orange[:,1])
    hold_slope = np.concatenate((hold_orange, slopes.reshape(slopes.shape[0], 1)), axis=1)
    quads = np.apply_along_axis(find_quad, 1, hold_slope)
    hold_quad = np.concatenate((orig, quads.reshape(slopes.shape[0], 1)), axis=1)
    diffs = np.empty(hold_quad.shape[0])
    for i, row in enumerate(hold_quad):
        diffs[i] = classifier[row[0],row[1],row[2]]
    return fn(diffs)
```

```python
def softmax(diffs):
    distances = np.abs(diffs-overall_mean)
    weights = np.exp(distances)/sum(np.exp(distances))
    return (diffs*weights).sum()
```

```python
classify_climb(df.iloc[0], softmax)
```

```python
def evaluate(row):
    return row.grade, classify_climb(row, lambda x: x.mean())
```

```python
actual_vs_pred = df.apply(evaluate, axis=1, result_type='expand')
actual_vs_pred
```
