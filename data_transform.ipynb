```python
DIMS = (18,11)
```

```python
import numpy as np
climbs = np.loadtxt('datasets/climbs.txt')
climbs
```

```python
climbs.shape
```

```python
def flattenClimbs(data, rows):
    arrs = []
    for i in range(rows):
        arrs.append(data[range(i, data.shape[0], rows)])
    return(np.concatenate(arrs, axis=1))
```

```python
flattened = flattenClimbs(climbs, DIMS[0]).astype(int)
flattened
```

```python
flattened.shape
```

```python
import csv
f = open('datasets/flattenedClimbs.csv',"w")
f.truncate()
writer = csv.writer(f)
for row in flattened:
    writer.writerow(row)
f.close()
```

```python
import pandas as pd
cols = []
for i in range(DIMS[0]):
    for j in range(DIMS[1]):
        cols.append((i,j))
df = pd.DataFrame(flattened, columns=cols).astype(bool)
df
```

```python
df['holds'] = df.apply(lambda row: row[row].index, axis=1)
df
```
