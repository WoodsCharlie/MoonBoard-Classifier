```python
import pandas as pd
df = pd.read_csv('datasets/flattenedClimbs.csv', header=None).astype(bool)
df
```

```python
grades = pd.read_csv('datasets/grades.txt', names=['grade'])
grades
```

```python
from sklearn.model_selection import train_test_split
df_train, df_test, grades_train, grades_test = train_test_split(df, grades)
```

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(df_train, grades_train)
```

```python
grades_pred = clf.predict(df_test)
grades_pred
```

```python
clf.score(df_test, grades_test)
```
